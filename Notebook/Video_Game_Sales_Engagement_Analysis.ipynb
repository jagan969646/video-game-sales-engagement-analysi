```python
games_df['Rating'] = games_df['Rating'].fillna(games_df['Rating'].median())
games_df['Plays'] = games_df['Plays'].fillna(0)
games_df['Wishlist'] = games_df['Wishlist'].fillna(0)
games_df['Backlogs'] = games_df['Backlogs'].fillna(0)
```

```python
games_df.columns
```

```python
games_df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
games_df.columns = (
    games_df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

games_df.columns
```

```python
text_cols = ['title', 'team', 'genres']

for col in text_cols:
    games_df[col] = games_df[col].astype(str).str.strip().str.lower()
```

```python
games_df['rating'] = games_df['rating'].fillna(games_df['rating'].median())
games_df['plays'] = games_df['plays'].fillna(0)
games_df['wishlist'] = games_df['wishlist'].fillna(0)
games_df['backlogs'] = games_df['backlogs'].fillna(0)
```

```python
sales_df.columns
```

```python
sales_df.columns = (
    sales_df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

sales_df.columns
```

```python
text_cols = ['name', 'platform', 'genre', 'publisher']

for col in text_cols:
    sales_df[col] = sales_df[col].astype(str).str.strip().str.lower()

```

```python
sales_df['year'] = sales_df['year'].fillna(sales_df['year'].median())
sales_df['publisher'] = sales_df['publisher'].fillna('unknown')
```

```python
merged_df = pd.merge(
    games_df,
    sales_df,
    left_on='title',
    right_on='name',
    how='inner'
)
```

```python
merged_df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(x='rating', y='global_sales', data=merged_df)
plt.title("User Rating vs Global Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='wishlist', y='global_sales', data=merged_df)
plt.title("Wishlist vs Global Sales")
plt.show()
```
